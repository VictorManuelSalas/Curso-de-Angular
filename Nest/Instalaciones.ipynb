## Como instalar Nest JS
En la terminal ejecutamos el comando 
`npm install -g @nestjs/cli` ya cuando termine puedes generar un nuevo proyecto.

### Listado de comandos útiles de NestJS
 
    "Obtener versión": "nest --version",
    "Crear un nuevo proyecto": "nest new nombre-proyecto",
    "Generar un módulo": "nest generate module nombre-modulo",
    "Generar un controlador": "nest generate controller nombre-controlador",
    "Generar un servicio": "nest generate service nombre-servicio",
    "Generar un recurso completo": "nest generate resource nombre-recurso",
    "Iniciar servidor en desarrollo": "npm run start:dev",
    "Compilar proyecto": "npm run build",
    "Ejecutar servidor en producción": "npm run start:prod",
    "Ejecutar tests": "npm run test",
    "Ejecutar tests con cobertura": "npm run test:cov",
    "Listar todos los comandos disponibles": "nest --help"
